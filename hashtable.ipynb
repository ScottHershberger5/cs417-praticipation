```python
posts = [
    "I love learning Python with @code_master!",
    "Is @alice coming to the hackathon?",
    "Shoutout to @code_master for the help today.",
    "Hey @bob, did you see the post by @alice?",
    "Python dictionaries are awesome. @code_master agrees!"
]

mentions = {}

for post in posts:
  if "@" in post:
    post = post[:len(post) - 1]
    num_of_ats = post.count("@")
    if num_of_ats > 1:
      pass #havent implemented multiple @'s yet
    if " " in post[post.index("@"):]:
      len_before = len(post[:post.index("@")])
      username = post[post.index("@"):post[post.index("@"):].index(" ") + len_before]
    else:
      username = post[post.index("@"):]

    # adds it to the list or adds 1 is alr exists
    if username in mentions:
      mentions[username] += 1
    else:
      mentions[username] = 1

  print(mentions)
```

Dictonaries are so popular because they are very lightweight and fast, also its easier to remember a key to lookup then a number so they are more user friendly.
