```python
class Fraction:
  def __init__ (self,top,bottom):
    self.num = top
    self.den = bottom

  def divide(self):
    real_num = self.num / self.den
    return real_num
```

```python
f1 = Fraction(5,4)
f2 = Fraction(3,4)
```

```python
print(f1,f2)
```

```python
print(f1.divide)
```
